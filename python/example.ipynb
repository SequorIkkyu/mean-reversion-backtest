```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import backtest
from signal import zscore_signal
```

```python
df = pd.read_csv("../data/price.csv")
price = df["close"].values

plt.plot(price)
plt.title("Price Series")
plt.show()
```

```python
signals = zscore_signal(price, window=20, z_entry=2.0)

plt.plot(signals)
plt.title("Trading Signals")
plt.show()

signals[:30]
```

```python
engine = backtest.BacktestEngine(
    initial_capital=100_000,
    transaction_cost_pct=0.001
)

result = engine.run_backtest(
    prices=price.tolist(),
    signals=signals.tolist(),
    dt_in_years=1/252
)

result.sharpe_ratio, result.max_drawdown, result.total_return
```

```python
plt.plot(result.equity_curve)
plt.title("Equity Curve")
plt.xlabel("Time")
plt.ylabel("Equity")
plt.show()
```

```python
plt.plot(result.pnl)
plt.title("Per-step PnL")
plt.grid(True)
plt.show()
```

```python
print("Total Return:   ", result.total_return)
print("Max Drawdown:   ", result.max_drawdown)
print("Sharpe Ratio:   ", result.sharpe_ratio)
```
